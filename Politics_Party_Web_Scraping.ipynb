```python
print(len(data.text))
print(data.text)
```

```python
soup = BeautifulSoup(data.text, "lxml")
print(soup.prettify())
```

```python
politics_parties = soup.select('.mw-headline a')
cite = soup.select('.mw-headline sup a')
print(politics_parties[0].string)
print(cite[0].string)
```

```python
politics_parties = list(set(politics_parties)-set(cite))
```

```python
politics_parties_list = []
for party in politics_parties:
  print(party.string)
  politics_parties_list.append(party.string)
print(len(politics_parties_list))
```

```python
import pandas as pd

data = {'politics_party_name': politics_parties_list}

df = pd.DataFrame(data)

df.to_csv(r'politics_party_name.csv', index=False, header=True)

print(df)
```
